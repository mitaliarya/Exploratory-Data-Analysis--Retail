## Data Visualisation

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df.Sales,shade=True,color='blue')
sns.kdeplot(df.Profit,shade=True,color='red')
plt.xlim(-100,1000)
```

- Profit is more than sales but there are some areas where profit could be increased.

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));

sns.countplot(df['Category'],ax=axs[0][0])
axs[0][0].set_title('Category',fontsize=20)

sns.countplot(df['Segment'],ax=axs[0][1])
axs[0][1].set_title('Segment',fontsize=20)

sns.countplot(df['Ship Mode'],ax=axs[1][0])
axs[1][0].set_title('Ship Mode',fontsize=20)

sns.countplot(df['Region'],ax=axs[1][1])
axs[1][1].set_title('Region',fontsize=20)

plt.tight_layout()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.title('Count of State wise shipments')
plt.show()
```

- State wise California has highest sales.

- States like Texas, New York, etc are active in business.

- Few states like North Dakota, Wyoming and West Virginia have less or no sales at all.

- To generate more profit we have to increase our sales in Texas and New York where we have potential customers.

```python
plt.figure(figsize = (6,6))
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

- This proves Office supplies have highest demand and others are comparatively less.

```python
df.groupby('Region')['Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

```python
df.groupby('Region')['Sales'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Sales in different regions')
plt.show()
```

- West generates most of the profit followed by East.

- Central and Southern region are generating less profit as compared to the Eastern and Western Region.

- Increase the sales in Central and Southern region in order to generate more profit in these region.

```python
df.groupby('Ship Mode')['Sales'].agg(sum).plot(kind='bar')
plt.ylabel('Sales')
plt.title('Sales in different regions')
plt.show()
```

```python
df.groupby('Ship Mode')['Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

- Sales are more in Standard Class in comparision to others.
- Also, profit are also more in Standard Class in comparision to others.

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.title('Count of Sub-categories')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['Quantity'])
plt.xticks(rotation=90)
plt.title('Count of quantities often sell together')
plt.show()
```

- Often 2 or 3 items sell together in most of the cases.

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'])
```

- To keep the profit margins the percentage of discount should be less than 25%

```python
df.groupby(['Category'])[['Profit','Sales']].agg(sum).plot(kind='bar',figsize=(10,4), )
plt.ylabel('Profit and Sales')
plt.show()
```

- This proves that Furniture is generating very less profit.

```python
df.groupby('Sub-Category')[['Sales','Profit']].agg(sum).plot(kind='barh',figsize=(15, 8))
plt.ylabel('Sub-Category')
plt.xlabel('Sales and Profit')
plt.show()
```

- We have 3 weak areas Bookcases, Supplies and Tables as we are facing loss on the sales of these items. On the sales of tables we are facing highest loss.

```python
plt.figure(figsize=(9,5))
sns.countplot(x='Region',hue='Ship Mode',data=df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of ship mode by region and ship mode')
```

- Least sales are in Southern region, we can attract this region by offering more discount. Standard mode of shipment is opted by most of the customers in each region and Same day shipment is opted by less number of customers.

## Observations: 
- Higher number of buyers are from New York and California.
- Most customers buy 2 or 3 quantity only.
- Max discount give is 0 to 20%.
- There is no correlation between profit and Discount.
- Profits and sales are maximum in west region and minimum in south region.
- There is highest loss in selling tables.
- Sales and profit are more in Standard Class.

### Conclusion:
##### The weak areas where one can work to make more profit are:
- We should limit the sales of furniture and increase that of office supplies and technology as it is generating very less profit as compared to its sales.

- In the sub-categories we are facing huge loss on tables so that should be minimized.

- The sales and profit in Southern and Central region are less so we should give more incentives like discount in order to increase sales, hence profit will increase.

- After the highest sales in California we have high value of potential customers in Texas and New York so we should concentrate more on these states to generate more profit.

- Few states like North Dakota, Wyoming and West Virginia have less or no sales at all so we also need to focus on these states.

- The percentage of discount offered should not be more than 25%, otherwise loss is faced.
